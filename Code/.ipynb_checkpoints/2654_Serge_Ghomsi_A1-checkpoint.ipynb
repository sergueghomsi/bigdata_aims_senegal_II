<h3>The Dataset</h3>
<p>For this assignment we will be using the sonar_train.csv data set</p>
<p>This data set contains signals obtained from a variety of different aspect angles, spanning 90 degrees for mines and 180 degrees for rocks. Each sample is a set of 60 numbers in the range 0.0 to 1.0, where each number represents the energy within a particular frequency band (integrated over a certain period of time). The output target contains the letter R if the object is a rock and M if it is a mine (metal cylinder).</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import timeit
%matplotlib inline
```

```python
#Opening the sonar_train.csv file
data = pd.read_csv('sonar.csv')

###Display informations and the different column/features names
print data.info()
```

<p>In this data set each band is a feature and the target is the type of the object</p>
<h4> <u>QUESTION 1 :</u> How many samples and how many features do we have?</h4>
<p><i>we have 208 samples and 60 features </i></p>

```python
###using the train_test_split command split the data set in two : 'train' and 'test' with a ratio ~75%-25%
from sklearn.cross_validation import train_test_split

data_train_1,data_test_1 = train_test_split(data) # default parameters split with ratio 75%-25%

print "taining set:",np.shape(data_train_1)," and testint set:",np.shape(data_test_1)
```

```python
###Saving of the training and testing data in .csv file 

data_train_1.to_csv('data_train1.csv')
data_test_1.to_csv('data_test1.csv')
```

```python
data_train = pd.read_csv('data_train1.csv')
data_test = pd.read_csv('data_test1.csv')
data_train.drop(data_train.columns[[0]], axis=1, inplace=True)
data_test.drop(data_test.columns[[0]], axis=1, inplace=True)
data_test
```

```python
#The last column in our sets correspond to the target
###Count how many of 'R' and 'M' you have in total
n_R1,n_R2 = np.where(data_train['Type']=='R'),np.where(data_test['Type']=='R')
n_M1,n_M2 = np.where(data_train['Type']=='M'),np.where(data_test['Type']=='M')


print "the total number of 'R' is ",len(n_R1[0])+len(n_R2[0]),
print "and the total number of 'M' is ", len(n_M1[0])+len(n_M2[0]) 
```

```python
#In order to make targets more easily usable we will replace 'R' and 'M' by 0 and 1 respectively.
###Do it for both sets using the command : .replace()
data_train_new =data_train.replace('R', 0).replace('M',1)
data_test_new = data_test.replace('R', 0).replace('M',1)

target_train = data_train_new.pop('Type') 
target_test = data_test_new.pop('Type')
```

```python
### convert pandas format data in array 
data_train_array = data_train_new.values
data_test_array = data_test_new.values
target_train_array = target_train.values
target_test_array = target_test.values
```

<h3>K-Nearest Neighbors approach</h3>

<p>In this section we will apply the <b>K-Nearest Neighbors</b> technique to the data set</p>

```python
from sklearn.neighbors import KNeighborsClassifier
###Fille the brackets to use the brute force approach and 7 neighbors
knn = KNeighborsClassifier(n_neighbors=7,algorithm='brute')

### Train the Nearest Neighbor Classifier "knn" on the training set
knn.fit(data_train_array,target_train_array)
```

<h4> <u>QUESTION 2 :</u> Which kind of impact has the choice of the algorithm?</h4>
<p><i>The choice of the algorithm has impact on the <b>computational cost or complexity (time and resource consuming)</b>. For one case where the number of sample is N, and F the number of features, <b> the complexities are O[F N²] for Brute Force, O[F N log(N)]  for K-D Tree and variable (depends of structure of data) for Ball tree </b> </i></p>

```python
#At any time it is possible to recall the parameters using :
knn.get_params()
```

<p>We now want to test the accuracy of the nearest neighbor classifier</p>


```python
###Compute the target prediction for the elements of the test set and compare with their true value
type_prediction = knn.predict(data_test_array)
comp = type_prediction - target_test_array
idx_same = np.where(comp==0)
idx_diff= np.where(comp!=0)
index_train = data_train.index
index_test = data_test.index

print "the algorithm made",len(target_test_array[idx_diff]),"bad predictions over the test data" 

plt.figure(figsize=(10,5))
plt.ylim(-1.5,1.5)
plt.xlim(0,60)
plt.scatter(index_test[idx_same],comp[idx_same],c='b',marker='o',s=20, label = 'Good prediction')
plt.scatter(index_test[idx_diff],comp[idx_diff],c='r',marker='o',s=20, label = 'Bad prediction')
plt.title("Comparaison between Prediction and target",fontsize=15)
plt.xlabel("index of data",fontsize=13)
plt.ylabel("Prediction - target of data_test",fontsize=13)
plt.legend()
plt.grid()
```

```python
#scikit actually includes an accuracy evaluation function :
###Compute the accuracy using knn.score()

print "the accuracy of this prediction is :", knn.score(data_test_array,target_test_array)
```

<h4> <u>QUESTION 3 :</u> Is the accuracy acceptable?</h4>
<p><i>En regardant l'erreur (24%) sur l'algorithme de prediction on peut dire qu'elle est plus ou moins acceptable, mais ici le but de cet algorithme est de pouvoir identifier le matériaux (roche ou le métal) à partir de notre algorithme, donc aller forer un le sol parce qu'à l'algorithme nous a predit qu'on y pouvait trouver du Métal avec une erreur de 24% est un grand risque pour l'investisseur. Donc dans notre cas la précision de prediction (score) n'est pas acceptable! </i></p>

<p>In the following cell repeat the training and the test steps with different knn parameters and try to find a good combination.</p>

```python
### step to find the combinaison of good parameters 

neigh = range(1,52) #different values of k-neighbors
algo = ['brute','kd_tree','ball_tree'] #list of different algorithms
weig = ['uniform','distance'] #list of weights

knn_1 =[]
for w in weig:
    for a in algo:
        for n in neigh:
            knn_2 = KNeighborsClassifier(n_neighbors=n,algorithm=a,weights=w)
            knn_1.append(knn_2)

#trainnig and prediction
type_prediction_1 = []
comp_cost = []
for i in range(len(knn_1)):
    start_time = timeit.default_timer() #initial time in second
    knn_1[i].fit(data_train_array,target_train_array)
    elapsed = timeit.default_timer() - start_time #final titme in second
    comp_cost.append(elapsed)
    type_prediction_2 = knn_1[i].predict(data_test_array)
    type_prediction_1.append(type_prediction_2)

#compute of score
scores_1 = []
for j in range(len(knn_1)):
    scores_2=knn_1[j].score(data_test_array,target_test_array)
    scores_1.append(scores_2)
ind_scor = np.where(scores_1==np.max(scores_1))

#compute the ruuning time
comput_cost = []
ind_score = ind_scor[0]
for l in ind_score:
    comput_cost.append(comp_cost[l])
    #print knn_1[l],scores_1[l],comp_cost[l]

index_comput = np.where(comput_cost == np.min(comput_cost)) #index of minimal computatinal time
best_comp_cost = comp_cost[ind_score[index_comput][0]] #minimal computatinal time
best_knn_par = knn_1[ind_score[index_comput][0]] #best parameters of our KNN algorthim
best_score_knn = scores_1[ind_score[0]] # best scores for good paremeters

print " "
print "The best combiniason of parameter of knn algorithm taking into account the computational time is:"
print best_knn_par
print "with score of : ",best_score_knn
print " and computational time of : ",best_comp_cost
```

<h4> <u>QUESTION 4 :</u> What is the best accuracy you get? What can you conclude?</h4>
<p><i>The best accuracy is 0.846153846154. </i></p>
<p> <u> <b>Conclusion: </b> </u>  Avec les paramètres du premier algorithme (brute force et k = 7) on avait eu un score de 0.76. En entrainant l'algorithme avec differents paramètres on réussit avoir 12 combinaisons avec le même score et on retenu celui qui avait plus petit temps d'exécution parmis les 12. Ainsi on a eu cette combinaison suivante:</p>

<h3>Decision Tree</h3>

<p>Using the same sets we now investigate the efficiency of a <b>decision tree classifier<b> </p>

```python
from sklearn import tree
dtr = tree.DecisionTreeClassifier(max_depth = 5)

###train the decision tree classifier "dtr"
dtr.fit(data_train_array,target_train_array)
```

```python
#Same as previously we can display some informations
dtr.get_params()
```

<h4> <u>QUESTION 5 :</u> From the list above, give the keywords influencing the test performed at a decison node?</h4>
<p><i>We have: <b>min_samples_leaf which is the minimal size for split</b>: c'est le nombre d'élément necéssaire pour la création d'un noeud, plus il est grand moins on aura de noeud. <b> Criterion</b>: Il permet de decider de la création des sous-arbres sous un noeud ou encore de déclarer qu'un noeud doit être une feuille. <b>max_depth: </b> La profondeur de l'algorithme influence sur la décision d'un noeud, plus celle ci est grande plus on aura de noeuds  </i></p>

<p>In the follwing, using the work done previously test the accuracy of the decision tree classifier.</p>

```python
###Compute the target prediction for the elements of the test set and compare with their true value

dtr_prediction = dtr.predict(data_test_array)
comp_dtr = dtr_prediction - target_test_array
idn_same = np.where(comp_dtr==0)
idn_diff= np.where(comp_dtr!=0)

print "the algorithm made",len(target_test_array[idn_diff]),"bad predictions over the test data" 

plt.figure(figsize=(10,5))
plt.ylim(-1.5,1.5)
plt.xlim(0,60)
plt.scatter(index_test[idn_same],comp_dtr[idn_same],c='b',marker='o',s=20, label = 'Good prediction')
plt.scatter(index_test[idn_diff],comp_dtr[idn_diff],c='r',marker='o',s=20, label = 'Bad prediction')
plt.title("Comparaison between Prediction and target",fontsize=15)
plt.xlabel("index of data_test",fontsize=13)
plt.ylabel("Prediction - target of data",fontsize=13)
plt.legend()
plt.grid()
```

```python
###Compute the accuracy using dtr.score()
print "the accuracy of this dtre prediction is :", dtr.score(data_test_array,target_test_array)
```

<p>Hereafter, using different parameters for our decision tree, try to improve the accuracy</p>

<h4> <u>QUESTION 6 :</u> Explain which parameter(s) you plan to change and why?</h4>
<p><i> Je vais changer le paramètres suivants: <b> criterion :</b> parce qu'il influence sur la décision des noeud qui a une infuence sur le score <b> max_depth :</b> C'est le paramètre qui fixe la profondeur maximale de l'arbre dont le score en dépend. plus l'arbre est profond plus le score sera meilleur sauf en cas de overfliting  <b> min_samples_split :</b> C'est lui qui définit le nombre d'élément necessaire pour la création d'un noeud, le score va fortement dépendre de ce paramètre qui est très important.<b> splitter :</b> c'est ce paramètre qui guide du type de séparation qui peut être soit aléatoire soit "best", et qui a une influence sur le score. </i></p>


```python
### Using of different parameters and finding the first score.
crit = ['gini','entropy']
split = ['best','random']
depth = range(1,20,2)
samp_s = range(2,7)
samp_l = range(1,6)
feat = ['auto','sqrt','log2']

dtr_1 =[]
for c in crit:
    for s in split:
        for d in depth:
            for ss in samp_s:
                for sl in samp_l:
                    for f in feat:
                        dtr_2 = tree.DecisionTreeClassifier(criterion=c, max_depth=d,splitter=s,
                                                             min_samples_leaf=sl,max_features=f,min_samples_split=ss)
                        dtr_1.append(dtr_2)

dtr_prediction_1 = []
for i in range(len(dtr_1)):
    dtr_1[i].fit(data_train_array,target_train_array)
    dtr_prediction_2 = dtr_1[i].predict(data_test_array)
    dtr_prediction_1.append(dtr_prediction_2)

scor_1 = []
for j in range(len(dtr_1)):
    scor_2=dtr_1[j].score(data_test_array,target_test_array)
    scor_1.append(scor_2)
ind_sco = np.where(scor_1==np.max(scor_1))
ind_scores = ind_sco[0]

for l in ind_scores:
    print dtr_1[l],scor_1[l] 
```

 <p> <b>Algorithm with has the best score:</b>DecisionTreeClassifier(class_weight=None, criterion='entropy', max_depth=13,max_features='auto', max_leaf_nodes=None, min_samples_leaf=4, min_samples_split=5, min_weight_fraction_leaf=0.0,presort=False, random_state=None, splitter='best') 0.942307692308
